Problem 1

The Randomness of the second dataset is due to the function rpois which creates a poisson distribution with the specified mean and the amount of numbers required. It is observed from the plot by calculation that the mean always comes to the one provided which is 25.

```r
sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
sales2<-rpois(12,25)  # random numbers, Poisson distribution, mean at 34, 12 numbers
#png(filename="problem1.png")
par(bg="cornsilk")

plot(sales1, col="red", type="o", ylim=c(0,100), xlab="Month", ylab="Sales" )
title(main="Monthly Sales")

lines(sales2, type="o", pch=2, lty=2, col="black")
grid(nx=NA, ny=NULL)
legend("topleft", inset=0, c("Sales 1","Sales 2"), fill=c("blue","red"), horiz=TRUE)
#dev.off()
```

Problem 3

To analyse the plots, the lines dividing the group or the box are called quartiles and the middle line is called the median that tells us that half of the values are greater than this and half of them are lower. Box plots are used to show overall patterns of response for a group. They provide a useful way to visualise the range and other characteristics of responses for a large group. The length of a box tells how much variation is there in the values present.

We see that the median for Sales 1 is 34 meaning that half the sales in Sales 1 is less than 34 and half more than 34. It can also be interpreted that 50 percent of the sales lie between the inter quartile range of about 30-50.

Similarly, the median for Sales 2 is 45 meaning half the Sales 2 is less than 45 and half more than 45. It can also be interpreted that 50 percent of the sales lie between the inter quartile range of about 35-55.

```r
#png(filename="problem3.png")
fn<-boxplot(sales,col=c("orange","green"))$stats
text(1.61, fn[3,2], paste("
Median =", fn[3,2]), adj=0, cex=.7)
text(0.61, fn[3,1],paste("
Median =", fn[3,1]), adj=0, cex=.7)
grid(nx=NA, ny=NULL)
#dev.off()
```

Problem 6

In this problem we plot the different locations visited on the world map by using the Google geocoding API. Also, we plot them on a map of only USA with the states visited marked by a marker. The red dot marker represents the places on the map which have been visited. The marker can be changed using the pch parameter. 

```r
library("ggmap")
library("maptools")
library(maps)
register_google(key = "ENTER_API_KEY_HERE") 
visited <- c("SFO", "Chennai", "London", "Melbourne","Lima","Peru", "Johannesbury, SA")
ll.visited <- geocode(visited)
visit.x <- ll.visited$lon
visit.y <- ll.visited$lat
#png(filename="problem6.png")
map("world", fill=TRUE, col="white", bg="lightblue", ylim=c(-60, 90), mar=c(0,0,0,0))
points(visit.x,visit.y, col="red", pch=18)
#dev.off()
#devtools::install_github("dkahle/ggmap") Can be used if register_google isn't recognized
```

Another usage of the Google geocode API, shows the map of United States divided into states. The places visited have been marked.

```r
library("ggmap")
library("maptools")
library(maps)
visited <- c("SFO", "New York", "Buffalo","El Paso", "Dallas, TX")
ll.visited <- geocode(visited)
visit.x <- ll.visited$lon
visit.y <- ll.visited$lat
#png(filename="problem6USA.png")
map("state", fill=TRUE, col=rainbow(50), bg="lightblue", mar=c(0,0,0,0))
points(visit.x,visit.y, col="yellow", pch=10)
#dev.off()
```
