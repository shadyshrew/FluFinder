# Plot 1.1 - Influenza Positive Tests National Summary reported by Clinical Labs for 2018-2019 Season

```r
library(repr)
options(repr.plot.width=13, repr.plot.height=8)
#png(filename = "Plot1-1.png")
pos1 <- read.csv(file="C:\\Users\\Admin\\Desktop\\CourseWork\\Spring 2019\\DIC\\Lab1\\part2\\pos1.csv",skip = 1,header = TRUE)
#ylab = "Number of Positive Specimens",

par(mar=c(5,5,4,4),xpd = FALSE)
midpts <- barplot(pos1[1:18,]$TOTAL.A + pos1[1:18,]$TOTAL.B,
                  ylim = c(0,14000), cex.main = 1,
                  space = 0,
                  xlab = "Week" , 
                  xaxs = "i",yaxs = "i",
                  las = 2,
                  col = "yellow1",lwd = 2)
title(main =  " Influenza Positive Tests Reported to CDC by U.S. Clinical Laboratories, 
                  National Summary, 2018-2019 Season",cex.main = 1)
axis(side = 1,lwd = 2,lwd.ticks = 1, xaxs = "i", yaxs = "i", at = 0:18, labels = FALSE)
#text(x=midpts,y = -1, pos$Week, cex=0.8, srt=70, xpd=TRUE, pos = 1, offset = 1.7)
text(x=midpts,y = -1, paste(pos1[1:18,]$YEAR,pos1[1:18,]$WEEK,sep = ""), cex=0.9, srt=65, xpd=TRUE,adj = 1.2)
axis(side = 1,lwd = 2,  labels = FALSE,lwd.ticks = 1,at = c(0,10))

mtext(side = 2, line = 3.5, "Number of Positive Specimens")


par(new = T,lwd = 2)
barplot(pos1[1:18,]$TOTAL.B, 
        ylim = c(0,14000),
        xaxs = "i",yaxs = "i",
        las = 2,
        space = 0,
        col = "green4",lwd = 2)
#seq(0,35)
par(new = T)
plot(pos1[1:18,]$PERCENT.POSITIVE,lty = 1,type = "l" ,xlab=NA, ylab=NA, axes = F  ,ylim = c(0,35), yaxs = "i", lwd = 3,xaxs = "i")
axis(side = 4,lwd = 2)

par(new = T)
plot(pos1[1:18,]$PERCENT.A,lty = 2,type = "l" ,xlab=NA, ylab=NA, axes = F  ,ylim = c(0,35), yaxs = "i",xaxs = "i", lwd = 3, col = "orange1")


par(new = T)
plot(pos1[1:18,]$PERCENT.B,lty = 3,type = "l" ,xlab=NA, ylab=NA, axes = F  ,ylim = c(0,35), yaxs = "i",xaxs = "i", lwd = 3, col = "green2")

mtext(side = 4, line = 2.5, 'Percent Positive')

legend("topleft",inset = 0.02,
       legend=c("A", "B","Percent Positive","% Positive Flu A","% Positive Flu B"),
       lty=c(NA,NA,1,2,3),bty = "n", pch=c(15, 15,NA,NA,NA), col=c("yellow1", "green4","black","orange1","green2"),horiz = FALSE)
#dev.off()
```

The WHO and NREVSS report to CDC the total number of respiratory specimens tested for influenza and the number positive for influenza by virus type. In addition, public health laboratories also report the influenza A subtype (H1 or H3) and influenza B lineage information of the viruses they test and the age or age group of the persons from whom the specimens were collected.

In this plot, we compare the Number of detected Positive Specimens as reported by Clinical Labs per week starting at week 40 of 2018 to the week of Jan 28th 2019.

The yellow bars represent the specimens tested positive for type A flu and the green bars represent the specimen tested positive for type B flu. The type A flu bars are stacked over the type B flu bars. This is plotted with respect to the vertical axis on the left hand side.

The solid black line represents the total postive specimens as a percentage. This is plotted with respect to the vertical axis on the right hand side.
The dotted lines represent the percentage of type A and B flu respectively.

Implementation:-

1) We first set the margins for the plot using the parameter 'mar' and then also use 'options' in the repr library to set the height and width o the plot. We then read the .csv data file by skipping the header. The data is loaded on to a dataframe in R.


2) We plot the barplot with the sum of type A and type B flu and then just the type B flu so that the final plot is a stacked bar plot. To allow plotting over one graph we define the parameter new as True. We access only the required column from the data frame using dataframe$columnname

3) The barplots are created with the axis set as false. The axis are added later using the 'axis' function. Then the labels are printed at an angle using the 'text' function and paramters 'srt' for angle specification and 'xpd = TRUE' to allow plotting text outside the plot area.

4) The plot function is used to plot the dotted lines for the last 3 categories using the 'lty' paramter to set line type,'lwd' to set the line thickness and col to set color.

5) The legend function is used to plot the legend for the graph in the topleft position.

```r
#https://stackoverflow.com/questions/24441775/how-do-you-create-a-us-states-heatmap-based-on-some-values
library(repr)
options(repr.plot.width=10, repr.plot.height=8)
par(mar = c(1,1,1,1))

heat <- read.csv(file="C:\\Users\\Admin\\Desktop\\CourseWork\\Spring 2019\\DIC\\Lab1\\part2\\Heat-map.csv")
#heat <- read.csv(file="C:\\Users\\Admin\\Desktop\\CourseWork\\Spring 2019\\DIC\\Lab1\\part2\\heatweek9.csv")
heat$region <- tolower(heat$STATENAME)

library(fiftystater)
data("fifty_states")


library(ggplot2)
library(maps)
#states <- map_data("state")
fifty_states$region <- fifty_states$id
for(i in 1:length(heat$ACTIVITY.LEVEL)){
  st = toString(heat$ACTIVITY.LEVEL[i])
  num = as.numeric(substr(st,7,nchar(st)))
  heat$levs[i] <- num
}

my.df <- merge(fifty_states,heat, by="region", all.heat=T)
my.df <- my.df[order(my.df$order),]

#colours=rev(heat.colors(10))
ggplot(my.df, aes(x=long,y=lat,group=group))+labs(title="                                           2018-2019 Season Week 4 ending Jan 26th,2019")+
  geom_polygon(aes(fill=my.df$levs))+
  geom_path()+ xlab("Longitude")+ ylab("Latitude")+
 scale_fill_gradientn(colours=rev(c("red3","red2","firebrick1","orangered","orange1","orange2","yellow1","yellow2","chartreuse","chartreuse2","chartreuse4")),na.value="grey90",name="ILI Activity Level",
                       breaks=seq(1,10),
                       labels=c("Minimal","Minimal","Minimal","Low","Low","Moderate","Moderate","High","High","High"))+ 
  theme_bw() + theme(plot.title = element_text(size=14, face="bold"))+
  theme(panel.border = element_blank(), panel.grid.major = element_blank(),
        panel.grid.minor = element_blank(), axis.line = element_line(colour = "black"),  legend.text = element_text(size = 10) ) + theme_void()+
  annotate("text", label = "Alaska", x = -120, y = 32, size = 4, colour = "Black")+
  annotate("text", label = "Hawaii", x = -111, y = 30, size = 4, colour = "Black")+
  coord_map()+ guides(fill = guide_legend(title ="         ILI\nActivity Level")) +
  theme(
    legend.key.size = grid::unit(1,"lines"),
    legend.title = element_text(size = 11, face = "bold"),
    legend.text = element_text(size = 11)
  )

```

Data collected in ILINet are used to produce a measure of ILI activity by state. Activity levels are based on the percent of outpatient visits in a state due to ILI and are compared to the average percent of ILI visits that occur during weeks with little or no influenza virus circulation. Activity levels range from minimal, which would correspond to ILI activity from outpatient clinics being below, or only slightly above, the average, to high, which would correspond to ILI activity from outpatient clinics being much higher than average.

Implementation:-

1) The data is read as a dataframe.

2) We use the fiftystater library to get all the uited states, ggplot's map_data does not provide data about all the states.

3) We first create a new calumn called region in our read dataframe which contains the state names in lower case.

4) The fiftystater data is loaded, this dataframe contains all the latitudes and longitudes for different states in the united states.

5) The fiftystater data also has a column called region which contains the names of the states in their lower case.

6) We run a for loop to read the Activity level data and convert it to a usable form,ex. Level 1 is converted to the integer 1 and so on.

7) We then merge the fiftystater data with our cdc dataframe using the merge function using the column "region", this produces a dataframe with all the latitudes and longitudes along with their activity levels and state name. This data frame is ordered ince the order may have been changed during the merge.

8) Finally, we use the ggplot function to plot the map using the latitude and longitude aesthetics. The scale_fill_gradientn function is used to create a gradient of colors to represent the different heat levels. The breaks are given at different levels to represent the activity level, which may be Minimal(Level 1)l, Low, Moderate or High(Level 10).
